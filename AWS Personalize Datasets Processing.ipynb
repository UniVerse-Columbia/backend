```python
import pandas as pd
import json
import ast
```

```python
user_events = pd.read_csv('user-events_final.csv')
user_tags = pd.read_csv('user-tags-final.csv')
```

```python
user_events.head()
```

```python
interactions = user_events[['id', 'attendees', 'timestamp']]
interactions['attendees'] = interactions['attendees'].apply(ast.literal_eval)

interactions = interactions.explode('attendees')
interactions = interactions.rename(columns={'id' : 'ITEM_ID', 'attendees': 'USER_ID', 'timestamp' : 'TIMESTAMP'})
interactions = interactions.assign(EVENT_TYPE='attending')
interactions['TIMESTAMP'] = interactions['TIMESTAMP'].apply(lambda x : pd.to_datetime(x).timestamp())
interactions.head()
```

```python
interactions.to_csv('interactions.csv', index=False)
```

```python
events = user_events[['id', 'title', 'tags', 'description', 'timestamp']]
events['timestamp'] = events['timestamp'].apply(lambda x : pd.to_datetime(x).timestamp())
events = events.rename(columns={'id' : 'ITEM_ID', 'title': 'TITLE', 'tags' : 'TAGS', 'description' : 'DESCRIPTION', 'timestamp' : 'TIMESTAMP'})
events.head()
```

```python
events['TAGS'] = events['TAGS'].str.split(';')
events = events.explode('TAGS')
events.head()
```

```python
events.to_csv('events.csv', index=False)
```

```python
user_tags['tags'] = user_tags['tags'].apply(ast.literal_eval)
users = user_tags.explode('tags')
users.head()
```

```python
users.to_csv('users.csv', index=False)
```
