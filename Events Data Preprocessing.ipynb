```python
import pandas as pd
import numpy as np
import regex as re
```

```python
# Upload events.csv in the same directory\
baseFile = "eventsFeed"
events = pd.read_csv('eventsFeed1.csv')
for i in range(2,7):
  tmp = pd.read_csv('eventsFeed'+ str(i)+'.csv')
  events = pd.concat([events,tmp],ignore_index=True, sort=False)
```

```python
events.head()
```

```python
events.shape
```

```python
events.columns
```

```python
events = events[["summary", "startshortdate", "starttime", "locationaddress", "locationlink",
                 "contactname", "contactphone", "contactemail", "categories",
                 "description", "cost"]]
events.head()
```

```python
events['tags'] = events['categories'].str.split(',')
events_per_tag = events.explode('tags')
unique_tags = events_per_tag['tags'].value_counts()
print(unique_tags[15:45])
```

```python
good_tags = [ 'Online',
             'Morningside', 'Graduate School of Arts and Sciences', 'Seminar', 'Arts',
             'International and Public Affairs', 'Humanities', 'Social Sciences' ]
             
print(len(good_tags))
```

```python
events['filtered_tags'] = events['tags'].apply(lambda x : set(x).intersection(set(good_tags)) if set(x).intersection(set(good_tags)) else np.NaN)
events.head()
```

```python
print(events.shape)
events = events.dropna(subset=['filtered_tags'])
print(events.shape)
```

```python
events.head()
```

```python
events["userId"] = events["contactname"].apply(lambda x: re.sub(" ","_",x.lower()))
```

```python
events.head()
```

```python
events['filtered_tags'] = events['filtered_tags'].apply(lambda x:";".join(x))
```

```python
events.head()
```

```python
shortlisted_users = pd.read_csv('shortlisted_users.csv')
```

```python
shortlisted_users.head()
```

```python
shortlisted_events = shortlisted_users.merge(events, how = 'right', left_on='userid', right_on='userId')
shortlisted_events.head()
```

```python
shortlisted_events.shape
```

```python
shortlisted_events = shortlisted_events.drop(columns = ['email', 'userid', 'locationlink', 'contactname', 'contactphone', 
                                   'contactemail', 'categories', 'cost', 'tags'])
```

```python
import uuid
shortlisted_events['id'] = [str(uuid.uuid4()) for _ in range(len(shortlisted_events))]
shortlisted_events = shortlisted_events.rename(columns={'summary': 'title', 
                                                        'userId' : 'organizerId',
                                                        'locationaddress' : 'address', 
                                                        'filtered_tags' : 'tags'})
```

```python
shortlisted_events.head()
```

```python
shortlisted_events['timestamp'] = pd.to_datetime(shortlisted_events['startshortdate'] + ' ' + shortlisted_events['starttime'], format='%m/%d/%y %I:%M %p')
shortlisted_events['timestamp'] = shortlisted_events['timestamp'].dt.tz_localize('US/Eastern')
shortlisted_events['timestamp'] = shortlisted_events['timestamp'] + pd.DateOffset(months=2)
```

```python
shortlisted_events = shortlisted_events.drop(columns=['startshortdate', 'starttime'])
shortlisted_events.head()
```

```python
shortlisted_events.to_csv('shortlisted_events.csv', index=False)
```
